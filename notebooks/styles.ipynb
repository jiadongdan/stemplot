```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from stemplot import set_style, reset_style
```

### without set_style

```python
fig, ax = plt.subplots(1, 1, figsize=(3.6, 2.8))
x = np.linspace(0, 10, 500)
y = np.sin(x)
ax.plot(x, y)
```

### after set_style

```python
set_style()

fig, ax = plt.subplots(1, 1, figsize=(3.6, 2.8))
x = np.linspace(0, 10, 500)
y = np.sin(x)
ax.plot(x, y)
```

### after reset_style

```python
reset_style()

fig, ax = plt.subplots(1, 1, figsize=(3.6, 2.8))
x = np.linspace(0, 10, 500)
y = np.sin(x)
ax.plot(x, y)
```
